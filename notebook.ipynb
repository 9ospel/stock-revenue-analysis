## Tesla Analysis

Tesla is one of the fastest-growing electric vehicle manufacturers. This section explores Tesla's historical stock price performance and revenue growth.

```python
# --- TESLA STOCK DATA ---
tesla_data = fetch_stock_data("TSLA")

# --- TESLA REVENUE DATA ---
tesla_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
tesla_revenue = extract_revenue(tesla_url, "Tesla Quarterly Revenue")

# --- VISUALIZE TESLA ---
make_graph(tesla_data, tesla_revenue, "Tesla")
```

3. Revenue growth and stock price generally move in the same direction, showing market confidence in Tesla's financial performance.

4. Stock price spikes often occur ahead of revenue increases, indicating investor anticipation of future performance.

## GameStop Analysis

GameStop is a retail gaming company that gained global attention during the meme stock phenomenon. This section analyzes the relationship between GameStop's financial performance and stock behavior.

```python
# --- GAMESTOP STOCK DATA ---
gme_data = fetch_stock_data("GME")

# --- GAMESTOP REVENUE DATA ---
gme_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
gme_revenue = extract_revenue(gme_url, "GameStop Quarterly Revenue")

# --- VISUALIZE GAMESTOP ---
make_graph(gme_data, gme_revenue, "GameStop")
```
